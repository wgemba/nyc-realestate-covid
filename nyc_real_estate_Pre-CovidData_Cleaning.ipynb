## Add Week Number Class Attribute
### Assign each data point to a week number based on a 7 day week period, starting from the first date of entry

```python
start_date = date(2019, 4, 27) 
end_date = date(2020, 2, 28)

delta_date = end_date - start_date   # returns timedelta
uniqueDates = []
for i in range(delta_date.days + 1):
    day = start_date + timedelta(days=i)
    uniqueDates.append(day)
uniqueDates
```

```python
tempDates_df = pd.DataFrame(columns = ['DATE'])
tempDates_df['DATE'] = uniqueDates
tempDates_df
```

```python
lst1 = []
# Create a list of lists of 7 instances of the number n, where the n is of a value from 1 to 44 and when the count of n's in a sublist exceeds 7, a new list is formed and appended.
for n in range(1,45):
    if lst1.count(n) <= 7:
        mult = [n,n,n,n,n,n,n]
        lst1.append(mult)
    else:
        n = n + 1

print(lst1,'\n')

# Concat all lists in lst1 into one list, concat_lst1 
concat_lst1 = []
for i in lst1:
    concat_lst1 += i

print(concat_lst1, '\n')
print(len(concat_lst1))
```

```python
tempDates_df['WEEK #'] = concat_lst1
tempDates_df
```

```python
tempDates_df.info()
```

```python
tempDates_df['DATE'] = pd.to_datetime(tempDates_df['DATE'])
tempDates_df
```

```python
# Merge with main DF
prior44weeks = prior44weeks.merge(tempDates_df, on = 'DATE')
prior44weeks
```

```python
prior44weeks_copy = prior44weeks
prior44weeks_copy['MONTH'] = pd.DatetimeIndex(prior44weeks_copy['DATE']).month
prior44weeks_copy
```

```python
# Defines function that utilizes the month value to create a season class, <1,2,3,4> corresponding to Winter, Spring, Summer, Autumn; respectfully. 
def seasonal(month):
    if month in [12,1,2]:
        return 1
    if month in [3,4,5]:
        return 2
    if month in [6,7,8]:
        return 3
    if month in [9,10,11]:
        return 4
```

```python
prior44weeks_copy['SEASON'] = prior44weeks_copy['MONTH'].apply(seasonal)
prior44weeks_copy
```

```python
prior44weeks_copy = prior44weeks_copy.drop(columns = ['RESIDENTIAL UNITS','DATE'])
```

```python
column_names = ['BORO', 'ZIP CODE', 'LAND SQFT','GROSS SQFT', 'YEAR BUILT', 'WEEK #', 'SEASON', 'SALE PRICE']
prior44weeks_copy = prior44weeks_copy.reindex(columns=column_names)
prior44weeks_copy
```
